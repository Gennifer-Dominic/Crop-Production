```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
# Load Datasets
df_crop=pd.read_excel("D:\\Guvi mini proj\\Crop-Production\\FAOSTAT_data.xlsx")
df_crop 
```

```python
df_crop.drop("Note",axis=1,inplace=True)
df_crop.drop("Year Code",axis=1,inplace=True)
```

```python
df_crop = df_crop[df_crop['Value'].notna()]
```

```python
df = df_crop.pivot_table('Value',['Domain', 'Area', 'Item Code (CPC)', 'Item', 'Year'],'Element')
```

```python
df = df.reset_index()
```

```python
df.fillna(0,inplace=True)
```

```python
df.drop('Item Code (CPC)',axis=1,inplace=True)
```

```python
col=['Domain','Area', 'Item', 'Year','Area harvested','Laying','Milk Animals','Producing Animals/Slaughtered','Stocks','Yield',
    'Yield/Carcass Weight','Production']
df=df[col]
df
```

```python
df
```

```python
x=df.drop('Production',axis=1)
y=df['Production']
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import OneHotEncoder,StandardScaler
from sklearn.compose import ColumnTransformer
from sklearn.tree import DecisionTreeRegressor

x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python

oneHotEncoder=OneHotEncoder(drop='first',sparse_output=False)
scaler=StandardScaler()

preprocessor=ColumnTransformer(
    transformers=[
        ('oneHotEncoder', OneHotEncoder(drop='first', sparse_output=False), [0,1,2]),
        ('scaler',scaler,[4,5,6,7,8,9,10])
    ],
    remainder='passthrough'
)

x_train_dummy=preprocessor.fit_transform(x_train)
x_test_dummy=preprocessor.transform(x_test)
```

```python
#Training Models
from sklearn.linear_model import LinearRegression,Lasso,Ridge
from sklearn.neighbors import KNeighborsRegressor
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import mean_squared_error,r2_score,mean_absolute_error


models={
    'lr':LinearRegression(),
    'lss':Lasso(),
    'rs':Ridge(),
    'knr':KNeighborsRegressor(),
    'dtr':DecisionTreeRegressor()
}

for name,mod in models.items():
    mod.fit(x_train_dummy,y_train)
    y_predict=mod.predict(x_test_dummy)
    model_name=name
    mse=mean_squared_error(y_test, y_predict)
    mae=mean_absolute_error(y_test, y_predict)
    score=r2_score(y_test, y_predict)
    print(f"{name} MSE: {mean_squared_error(y_test,y_predict)} Score: {r2_score(y_test,y_predict)}")
    #insertMeasure(model_name,mse,score,mae)

y_predict
```
